# Using Haar Cascade Pretrained Face Detector

### Importing Libraries

```python
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
```

### Defining the labels

```python
subjects = ["", "Blood", "No_Blood"]
```

### Returning the detected face.

```python
def detect_face(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
    faces = face_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5);
    if (len(faces) == 0):
        return None, None
    x, y, w, h = faces[0]
    
    return gray[y:y+w, x:x+h], faces[0]
```

## Data Preparation

### Data Directory should be like 

dataset
----------b1
           blood_images.png
           .
           .
           .
----------b2
           no_blood_images.png
           .
           .
           .

```python
def prepare_training_data(data_folder_path):
    dirs = os.listdir(data_folder_path)
    faces = []
    labels = []
    
    for dir_name in dirs:
        if not dir_name.startswith("b"):       #To eliminate other directories
            continue;
        label = int(dir_name.replace("b", ""))   # b1 for blood images and b2 for noblood images
        subject_dir_path = data_folder_path + "/" + dir_name
        subject_images_names = os.listdir(subject_dir_path)
        
        for image_name in subject_images_names:
            if image_name.startswith("."):        #to eliminate .DStore file
                continue;
            image_path = subject_dir_path + "/" + image_name
            image = cv2.imread(image_path)
            
            face, rect = detect_face(image)
            
            if face is not None:           # Appending it into lists
                faces.append(face)
                labels.append(label)
                
    return faces, labels
```

```python
faces, labels = prepare_training_data("dataset_hc/training_data/")

#print total faces and labels
print("Total faces: ", len(faces))
print("Total labels: ", len(labels))
```

### Defining the LBPH algorithm for face recognition

```python
face_recognizer = cv2.face.LBPHFaceRecognizer_create()
```

### Training the face dataset on LBPH Face recognition algorithm

```python
face_recognizer.train(faces, np.array(labels))
```

### Defining function for drawing rectangle and put text on the co-ordinates.

```python
def draw_rectangle(img, rect):
    (x, y, w, h) = rect
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)

def draw_text(img, text, x, y):
    cv2.putText(img, text, (x, y), cv2.FONT_HERSHEY_PLAIN, 1.5, (0, 255, 0), 2)
```

# Function for the prediction of images

```python
def predict(test_img):
    img = test_img.copy()
    face, rect = detect_face(img)
    label = face_recognizer.predict(face)
    label_text = subjects[label[0]]
    draw_rectangle(img, rect)
    draw_text(img, label_text, rect[0], rect[1]-5)
    return img
```

### Testing is done

```python
test_img1 = cv2.imread("dataset_hc/training_data/b1/blood_1.jpg")
test_img2 = cv2.imread("dataset_hc/training_data/b2/noblood_17.jpg")

#perform a prediction
predicted_img1 = predict(test_img1)
predicted_img2 = predict(test_img2)

#display both images
cv2.imshow(subjects[1], predicted_img1)
cv2.imshow(subjects[2], predicted_img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
